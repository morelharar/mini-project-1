```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#test the numpy 
x=np.linspace(0,2*np.pi,1000)
y=np.sin(x)
#find the max and min points
max_y=np.max(y)
min_y=np.min(y)
#find corresponding x values for max and min
max_x=x[np.argmax(y)]
min_x=x[np.argmin(y)]
#plot the sin wave
plt.plot(x,y,label='sin(x)',color='b')
#mark the max and min points
plt.scatter(max_x,max_y,color='red',label=f'max:({max_x:.2f},{max_y})')
plt.scatter(min_x,min_y,color='green',label=f'min:({min_y:.2f},{min_y})')
#add labels and title
plt.title('sin Function with the max and min points')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
#show the plot
plt.grid(True)
plt.show()
```

```python
#Part 2-Q1 
# Paramenters
frequency = 5  # Frequency in Hz
sampling_interval = 0.01  # Sampling in sec
original_duration = 2  # Duration of the signal in sec

# sine wave
time_original = np.arange(0, original_duration, sampling_interval)
signal_original = np.sin(2 * np.pi * frequency * time_original)

# Downsample the signal to 20% of its original length
new_length = int(len(signal_original) * 0.2)
signal_downsampled = signal_original[::int(len(signal_original) / new_length)]

# Find the max and min points of the original sine wave
max_point = np.max(signal_original)
min_point = np.min(signal_original)

# Output the results
print("Max point:", max_point)
print("Min point:", min_point)

# Find the time corresponding to max and min points
max_time=time_original[np.argmax(signal_original)]
min_time=time_original[np.argmin(signal_original)]

# Output the results
print("Max point:Amplitude=",max_point, "at time=",max_time)
print("Min point:Amplitude=",min_point,"at time=",min_point)


# Plot the original and downsampled signals
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(time_original, signal_original, label="Original Signal")
plt.scatter(max_time,max_point,color='g',label="Max Point")
plt.scatter(min_time,min_point,color='b',label="Min Point")
plt.text(max_time,max_point,f"({max_time:.2f},{max_point:.2f})",fontsize=10, color='g')
plt.text(min_time,min_point,f"({min_time:.2f},{min_point:.2f})",fontsize=10,color='b')
plt.title("Original Sine Wave")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.legend()

time_downsampled = np.linspace(0, original_duration, new_length)
plt.subplot(2, 1, 2)
plt.plot(time_downsampled, signal_downsampled, label="Downsampled Signal", color='r')
plt.title("Downsampled Sine Wave (20% Length)")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")

#mark the max and min points

plt.tight_layout()

plt.show()
```

```python
# part 2- Q2 laptop price- dataset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1- Load the dataset
file_path=r"C:\Users\morel\OneDrive\ \  -    \- \Projects\Mini project 1\laptop_price - dataset.csv"
df=pd.read_csv(file_path)

# Check the diatels from the dataset
print(df.head())

print(df.columns)
print(df.info())
df.columns = df.columns.str.strip()

```

```python
# 2- Plot the distribution of laptop prices
plt.figure(figsize=(10,6))
sns.histplot(df['Price (Euro)'],kde=True,bins=30,color='blue')
plt.title('Price distribution of laptops')
plt.xlabel('Price (in Euro)')
plt.ylabel('Frequency')
plt.show()

```

```python
# 3-Finding the Company with the Most Expensive Laptop on average
average_price_per_company=df.groupby('Company')['Price (Euro)'].mean()
# Displying the Company with the most expensive laptop (avarage)
most_expensive_company=average_price_per_company.idxmax()
most_expensive_avg_price=average_price_per_company.max()

print(f"Company with the most expensive laptops on average:{most_expensive_company}")
print(f"Average price for {most_expensive_company}: ${most_expensive_avg_price}")

# Plot the average ptice 
plt.figure(figsize=(12,6))
average_price_per_company.sort_values(ascending=False).plot(kind='bar', color='green')
plt.title('Average laptop price by company')
plt.xlabel('Company')
plt.ylabel('Average Price (in Euro)')
plt.xticks(rotation=45)
plt.show()
```

```python
# 4- check the unique values in 'OpSys' column
print(df['OpSys'].unique())
# Standarize the 'opSys' column 
df['OpSys'].str.strip().str.lower()
```

```python
# replace known variation with standarized values
df['OpSys']=df['OpSys'].replace({
    'windows 10':'windows',
    'win 10':'windows',
    'mac os':'macos',
    'linux ubuntu':'ubuntu',
    'ubuntu os':'ubuntu'
})
# Checking 
print(df['OpSys'].unique())
```

```python
# 5- Plot the distribution of prices
os_list=df['OpSys'].unique()
for os in os_list:
    plt.figure(figsize=(10,6))
    sns.histplot(df[df['OpSys']==os]['Price (Euro)'],kde=True,bins=30)
    plt.title(f'Price distribution fo {os.capitalize()}')
    plt.xlabel('Price(in Euro)')
    plt.ylabel('Frequency')
    plt.show()
```

```python
# 6- the relationship between RAM and Computer Price 
# RAM information should be imported from the 'Memory' column
df['RAM']=df['Memory'].str.extract('(\\d+)',expand=False).astype(float)
# Plot the relatiion between RAM and Price
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='RAM',y='Price (Euro)')
plt.title('The relationship between RAM and laptop price')
plt.xlabel('RAM (in GB)')
plt.ylabel('Price (in euro)')
plt.show()
```

```python
# 7- Creating a new column for storage type 
# importing storage type from 'Memory' column
df['Storage type']=df['Memory'].str.extract('(SSD|HDD)',expand=False)

# Display the first few rows in order to verify the new column 
print(df[['Memory', 'Storage type']].head())
```

```python
# Bonus Q-
# 1- What is the price difference between the laptops with SSD HDD storage?
# Before we must ensurte that the 'Storage type' column exists (I've already created)
if 'Storage type' not in df.columns:
    df['Storage type']=df['Memory'].str.extract('(SSD|HDD)',expand=False)

# in order to compare and calculate the price difference we must ensure that the 'Price (Euro)' column contains numeric data.

df['Price (Euro)']=pd.to_numeric(df['Price (Euro)'], errors='coerce')

# Calculate the average price # and we must filter the DataFrame 
avg_price_ssd=df[df['Storage type']=='SSD']['Price (Euro)'].mean() # this function filters rows with dtorage type 'SSD'
avg_price_hdd=df[df['Storage type']=='HDD']['Price (Euro)'].mean()

# the difference between the laptops type prices
price_difference=avg_price_ssd-avg_price_hdd

# Print the outputs 
print(f"SSD laptops average price:{avg_price_ssd}")
print(f"HDD laptops average price:{avg_price_hdd}")
print(f"Price difference:{price_difference}")
```

```python
# Bonus Q-
# 2- how does storage size affect laptop price?
df['Storage size']=df['Memory'].str.extract(r'(\d+)',expand=False)

# Calulate the avg price for each storage size
storage_size_price=df.groupby('Storage size')['Price (Euro)'].mean()
print(storage_size_price)

# Plot the results
plt.figure(figsize=(10,6))
sns.lineplot(x=storage_size_price.index, y=storage_size_price.values, marker='o')
plt.title('Average price by store size')
plt.xlabel('Storage size (in GB)')
plt.ylabel('Average price (in Euro)')
plt.grid(True)
plt.show()
```
